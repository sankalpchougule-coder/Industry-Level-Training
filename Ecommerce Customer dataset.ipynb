```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ans

df = pd.read_csv('ecommerce_customer_churn_dataset.csv')
df
```

```python
df.head()
df.tail()
df.shape
df.info()
df.describe()
```

```python
df.describe().T
```

```python
cat_col = df.select_dtypes(include = ['object']).columns.tolist()
num_cols = df.select_dtypes(include = ['int64', 'float64']).columns.tolist()

print("Categorical columns:", cat_col)
print("Numeric columns:", num_cols)

for col in cat_col:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
df['Age'] = df ['Age'].clip (lower=18, upper=90)
df['cart_Abandonment_Rate'] = df['Cart_Abandonment_Rate'].clip(upper=100)
df['Total_Purchases'] = df['Total_Purchases'].clip(lower=0)
print("Erroneous values corrected for Age, Cart_Abandonment_Rate, and Total_Purchases.")
```

```python
numerical_cols = df.select_dtypes (include=['float64', 'int64']).columns
for col in numerical_cols:
    if col != 'Payment_Method_Diversity': # Payment_Method_Diversity will be handled separately
        df[col].fillna(df[col].median(), inplace=True)
print("Missing values in numerical columns (except Payment_Method_Diversity) imputed with their median.")
```

```python
numerical_cols = df.select_dtypes (include=['float64', 'int64']).columns
for col in numerical_cols:
      if col != 'Payment_Method_Diversity': # Payment_Method_Diversity will be handled separately
            df[col] = df[col].fillna(df[col].median())
print("Missing values in numerical columns (except Payment_Method_Diversity) imputed with their median.")
```

```python
df['Payment_Method_Diversity'].fillna (df['Payment_Method_Diversity'].mode() [0], inplace=True)
print("Missing values in 'Payment_Method_Diversity' imputed with its mode.")
```

```python
df['Payment_Method_Diversity'] = df[ 'Payment_Method_Diversity'].fillna(df['Payment_Method_Diversity'].mode()[0])
print("Missing values in 'Payment_Method_Diversity' imputed with its mode.")
```

```python
print("Missing values after imputation:")
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna('0')
```

```python
df_filled = df.fillna(df.mean(numeric_only=True))

print("\nMissing values after:")
print(df_filled.isnull().sum())

df_filled.to_csv("ecommerce_customer_churn_dataset.csv",index=False)

df_filled.head()
```
