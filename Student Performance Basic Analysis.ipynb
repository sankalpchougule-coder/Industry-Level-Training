This Project includes the Basic Data Analysis of student performance dataset

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Student_Performance.csv')
df
```

```python
# Basic operations to perform
df.head()
df.tail()
df.info()
df.shape
df.describe()
```

```python
df.describe().T  #Transpose
```

```python
# Check basic info for categorical columns

cat_col = df.select_dtypes(include = ['object']).columns.tolist()
num_cols = df.select_dtypes(include = ['int64', 'float64']).columns.tolist()

print("Categorical columns:", cat_col)
print("Numeric columns:", num_cols)

for col in cat_col:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts())
```

```python
# Check missing values

df.isnull().sum()
```

```python
# Check for duplicate rows

print("Number of duplicate rows:",df.duplicated().sum())
```

```python
#Drop duplicates if any

df = df.drop_duplicates().reset_index(drop=True)
print("Shape after dropping duplicates:", df.shape)
```

```python
#Check of duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Basic sanity checks
#Age range, attendance ,scores sanity

print("Age range:", df['age'].min(), "to", df['age'].max())
print("Attendance range:", df['attendance_percentage'].min(), "to", df['attendance_percentage'].max())
print("Overall score range:", df['overall_score'].min(), "to", df['overall_score'].max())
```
