This Project includes the Basic Data Analysis of Airline delay dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Airline_Delay_Cause.csv')
df
```

```python
df.head()
df.tail()
df.info()
df.shape
df.describe()
```

```python
df.describe().T
```

```python
cat_cols = df.select_dtypes(include = ['object']).columns.tolist()
num_cols = df.select_dtypes(include = ['int64' , 'float64']).columns.tolist()

print("Categorical columns:", cat_cols)
print("Numeric columns:", num_cols)

for col in cat_cols:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
df['carrier'] = df['carrier'].fillna('NA')
```

```python
df['carrier_name'] = df['carrier_name'].fillna('NA')
```

```python
df['airport'] = df['airport'].fillna('NA')
```

```python
print(df.isnull().sum())
```

```python
df_filled = df.fillna(df.mean(numeric_only = True))

print("\nMissing values after:")
print(df_filled.isnull().sum()) 

df_filled.to_csv("operations.csv",index=False)
df_filled.head()
```
